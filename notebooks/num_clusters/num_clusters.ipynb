# Determine the number of clusters present in the sequence embeddings

Use the Calinski Harabasz index for 5,000 sequences embedded with the Bacillus carbohydrate metabolism trained Protvec model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
ch_data = pd.read_csv('calinski_bacil.csv', index_col = 'Unnamed: 0')
wss_data = pd.read_csv('wss_bacil.csv', index_col = 'Unnamed: 0')
```

```python
#Calinski Harabasz Index 
#set the number of clusters 
c = len(ch_data.columns)

labels = [i+2 for i in range(c)]
ch_data.columns = labels
ch_data_df = ch_data.T.unstack().to_frame().sort_index(level=1).reset_index()


sns.set(font_scale = 3.5)
fig, ax = plt.subplots(1, 1, figsize = (25,15))
sns.boxplot(x='level_1', y=0, data=ch_data_df, color = '#FFB944')
#ax.set_facecolor('white')
plt.xlabel('Number of Clusters', fontsize = 40)
plt.ylabel('Calinski-Harabasz index', fontsize = 40)
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
plt.savefig("CHindex_bacillus_embeddedbacil.png")

#Within Clusster Sum of Squares
c = len(wss_data.columns)

labels = [i+2 for i in range(c)]
wss_data.columns = labels
wss_data_df = wss_data.T.unstack().to_frame().sort_index(level=1).reset_index()

sns.set(font_scale = 3)
fig, ax = plt.subplots(1, 1, figsize = (25,15))
sns.boxplot(x='level_1', y=0, data=wss_data_df, color = '#FFB944')
plt.xlabel('Number of Clusters', fontsize = 40)
plt.ylabel('Within Cluster Sum of Squares', fontsize = 40)
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
plt.savefig("wss_bacillus_embeddedbacil.png")
```
